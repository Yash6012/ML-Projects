```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Algerian_forest_fires_cleaned_dataset.csv')
```

```python
df.head()
```

```python
df.drop(['month', 'day', 'year'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Classes'].value_counts()
```

```python
df['Classes'].unique()
```

```python
# Encoding
df['Classes'] = np.where(df['Classes'].str.contains('not fire'), 0, 1)
```

```python
df['Classes'].unique()
```

```python
# Dependent and independent features
X = df.drop('FWI', axis=1)
y = df['FWI']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
X_train.shape, X_test.shape
```

```python
# Feature Selection
# Highh correlation -> Similar features
X_train.corr()
```

## Multicollinearity 
Multicollinearity is a statistical phenomenon that occurs in Multiple Regression when two or more independent variables (features) are highly correlated with each other.

So we can drop 1 of the independent features.

```python
# Check for multicollinearity
plt.figure(figsize=(12,10))
corr = X_train.corr()
sns.heatmap((corr), annot=True)
```

```python
def correlation(dataset, threshold):
    col_corr = set()
    corr_matrix = dataset. corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold:
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
# threshold -- Domain Expertise
corr_feature = correlation(X_train, 0.85)
```

```python
# Drop features when correlation is more than 0.85
X_train.drop(corr_feature, axis=1, inplace=True)
X_test.drop(corr_feature, axis=1, inplace=True)
X_train.shape, X_test.shape
```

# Feature Scaling or Standardization

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## Box Plots to understand effects of Standard Scaler

```python
plt.subplots(figsize=(15, 5))
plt.subplot(1, 2, 1)
sns.boxplot (data=X_train)
plt.title('Before Scaling')
plt.subplot(1, 2, 2)
sns.boxplot(data=X_train_scaled)
plt.title('X_train After Scaling')
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
regression = LinearRegression()
regression.fit(X_train_scaled, y_train) # Gives a best fit line
y_pred = regression.predict(X_test_scaled) # give predicted values
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
print("MSE: ", mse)
print("MAE: ", mae)
print("R2: ", r2)
plt.scatter(y_test, y_pred)
```
